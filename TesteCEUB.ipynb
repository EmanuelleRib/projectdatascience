```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

np.random.seed(42)


num          = 100
idade        = np.random.randint(18, 80, size=num)
altura_media = np.random.normal(loc=170, scale=10, size=num)
peso_media   = np.random.normal(loc=70, scale=15, size=num)

df = pd.DataFrame({
    'Idade': idade,
    'Altura (cm)': altura_media,
    'Peso (kg)': peso_media
})

df = df[(df['Altura (cm)'] >= 150) & (df['Altura (cm)'] <= 200)]
df = df[(df['Peso (kg)'] >= 40) & (df['Peso (kg)'] <= 120)]


df['IMC'] = df['Peso (kg)'] / ((df['Altura (cm)'] / 100) ** 2)
df.fillna(df.mean(), inplace=True)
sns.set(style="whitegrid")

sns.histplot(df['Idade'], bins=15, kde=True)
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Altura (cm)', y='Peso (kg)', data=df)
plt.title('Altura vs Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(df['IMC'], bins=15, kde=True)
plt.title('Distribuição de IMC')
plt.xlabel('IMC')
plt.ylabel('Frequência')
plt.show()
```
